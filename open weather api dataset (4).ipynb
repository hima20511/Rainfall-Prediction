```python
import requests
from datetime import datetime
import pandas as pd

api_key = '8c73b293dae4e3bf46baa72cbab742fd'

cities = ['Srikakulam','Visakhapatnam','Rajahmundry','vijayawada','Tirupati','Nellore','Kurnool',
          'Kadapa','Ongole','Anantapur','Chittoor','Vizianagaram','Bhimavaram','Narasaraopet','Mumbai',
          'Hyderabad','Pune','Chennai','Bengaluru','Kolkata','Jaipur','Lucknow','Surat','Chandigarh',
          'Delhi','Nagpur','Mysore','Patna','Tirunelveli','Udaipur']
data = []

for location in cities:
    forecast_endpoint = f'http://api.openweathermap.org/data/2.5/forecast?q={location}&appid={api_key}'
    forecast_data = requests.get(forecast_endpoint).json()
    
    
    for i in range(0, 40):
        date = datetime.fromtimestamp(forecast_data['list'][i]['dt']).strftime('%Y-%m-%d %H:%M:%S')
        city = forecast_data['city']['name']
        min_temp = forecast_data['list'][i]['main']['temp_min']
        max_temp = forecast_data['list'][i]['main']['temp_max']
        wind_direction = forecast_data['list'][i]['main']['sea_level']
        wind_speed = forecast_data['list'][i]['main']['grnd_level']
        humidity = forecast_data['list'][i]['main']['humidity']
        pressure = forecast_data['list'][i]['main']['pressure']
        cloud = forecast_data['list'][i]['clouds']['all']
        temperature = forecast_data['list'][i]['main']['temp']
        
        # Placeholder for rain conditional statement
        if 'rain' in forecast_data['list'][i]:
            rain = 1
        else:
            rain = 0
        
        # Placeholder for trm_rain conditional statement
        if i + 8 >= 40:
            trm_rain = 0
        elif 'rain' in forecast_data['list'][i + 8]:
            trm_rain = 1
        else:
            trm_rain = 0
        
        data.append({'Date': date, 'City': city, 'Min Temp': min_temp, 'Max Temp': max_temp,
                     'Wind Direction': wind_direction, 'Wind Speed': wind_speed, 'Humidity': humidity,
                     'Pressure': pressure, 'Cloud': cloud, 'Temperature': temperature, 'Rain': rain,
                     'Trm Rain': trm_rain})

df = pd.DataFrame(data)
print(df)
```
